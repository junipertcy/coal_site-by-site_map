```python
import numpy as np
import pandas as pd
csv_d = pd.read_csv("coal_plant_info.csv")
```

```python
plants = {
    "type": "FeatureCollection",
    "features": []
}
for idx, _id in enumerate(csv_d["ID"]):
    plants["features"] += [{
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [float(csv_d["Lon"][idx]), float(csv_d["Lat"][idx])]
        },
        "properties": {
            "id": str(csv_d["ID"][idx]),
            "name_en": str(csv_d["Source.Name"][idx]),
            "name": str(csv_d["Plant name"][idx]),
            "operator": str(csv_d["Operator"][idx]),
            "investors": str(csv_d["Operator owners / investors"][idx]),
            "fuels_used": str(csv_d["Fuels used"][idx]),
            "capacity": str(csv_d["Capacity of plant"][idx]),
            "status": str(csv_d["Status of project "][idx]).replace("\n", ""),
            "pollutants": {
                "NO": float(csv_d["NO"][idx]),
                "NO2": float(csv_d["NO2"][idx]),
                "HNO3": float(csv_d["HNO3"][idx]),
                "NO3": float(csv_d["NO3"][idx]),
                "PM15": float(csv_d["PM15"][idx]),
                "PM10": float(csv_d["PM10"][idx]),
                "PPM25": float(csv_d["PPM25"][idx]),
                "Hg0": float(csv_d["Hg0"][idx]),
                "RGM": float(csv_d["RGM"][idx]),
                "Hgp": float(csv_d["Hgp"][idx])
            }
        }
    }]
```
